```python
# 종합 프로젝트용 새 데이터 생성 — 베이커리 체인 "빵셀러" 운영 데이터
np.random.seed(13)
n_days = 90
n_stores = 6
n_rows = n_days * n_stores * 4   # 매장 x 일자 x 시간대(4구간)

stores = [f"S{i:02d}" for i in range(1, n_stores + 1)]
items = ["크루아상", "식빵", "케이크", "샌드위치", "쿠키"]

bakery = pd.DataFrame({
    "store_id": np.tile(stores, n_rows // n_stores)[:n_rows],
    "date_str": np.random.choice([
        "2025-04-01", "2025/04/01", "20250401",
        "2025-05-15", "2025/05/15", "20250515",
        "2025-06-20", "2025/06/20", "20250620",
    ], n_rows),
    "item": np.random.choice(items, n_rows),
    "quantity": np.random.choice([1, 1, 2, 2, 3, 5, 50], n_rows),  # 50은 의심값
    "unit_price": np.random.choice([3500, 4500, 5500, 8500, 12000], n_rows),
})
bakery["revenue"] = bakery["quantity"] * bakery["unit_price"]

# 오염 심기
bakery.loc[np.random.choice(n_rows, 60, replace=False), "revenue"] = np.nan
bakery.loc[5, "revenue"] = -45000  # 환불 또는 실수
bakery.loc[bakery.sample(10, random_state=1).index, "store_id"] = " S01 "  # 공백
bakery.loc[bakery.sample(8, random_state=2).index, "item"] = "케익"        # 표기 혼재('케이크' vs '케익')
bakery = pd.concat([bakery, bakery.iloc[[0, 1, 2, 3]]], ignore_index=True)   # 중복 4건

print("빵셀러 데이터 생성 완료:", bakery.shape)
bakery.head()
```

```python
# 원본 copy 함수
def bakery_copy(bakery):
    bakery = bakery.copy()
    return bakery

bakery_clean = bakery_copy(bakery)
bakery_clean.head()
```

```python
#결측 확인
bakery_clean[["store_id", "date_str", "item", "quantity", "unit_price", "revenue"]].isna().sum()
```

```python
#컬럼별 value_counts()
columns_to_check = ["date_str", "item", "quantity", "unit_price", "revenue"]   # 확인하고 싶은 컬럼들

for col in columns_to_check:
    print(f"=== {col} ===")
    print(bakery_clean[col].value_counts(dropna = False))
    print()
```

```python
#item별 unit_price 확인

bakery_clean[bakery_clean["item"] == "케익"]["unit_price"].value_counts()
```

```python
#revenue 금액 검증 (quantity * unit_price)

subset = bakery_clean[bakery_clean["quantity"] > 0].copy()

# 직접 계산한 예상 매출
subset["expected_revenue"] = subset["unit_price"] * subset["quantity"]

# 실제 revenue와 차이나는 행만 필터링(na 포함)
mismatch = subset[subset["expected_revenue"] != subset["revenue"]]
print(mismatch)

# na 미포함 리스트
mismatch = bakery_clean[
    bakery_clean["revenue"].notna() & (subset["expected_revenue"] != bakery_clean["revenue"])
]
print(mismatch)
```

```python
#이상치 처리
#revenue 음수 결측으로 처리
def revenue_clean(bakery):
    bakery["revenue"] = bakery["revenue"].where(
        (bakery["revenue"] >= 0), 
        np.nan)
    bakery["revenue"] = bakery["revenue"].astype("Int64")
    return bakery

#revenue 결측 계산값(unit_price * quantity)처리
def revenue_na_clean(bakery):
    bakery["revenue_was_missing"] = bakery["revenue"].isna()   # 플래그 먼저 남기기
    bakery["revenue"] = bakery["revenue"].fillna(
        bakery["unit_price"] * bakery["quantity"]
    )
    return bakery

#날짜 파싱
def date_str_parsing(bakery):
    bakery["date_str"] = pd.to_datetime(bakery["date_str"], format = "mixed")
    return bakery

#문자열 처리
def store_id_clean(bakery):
    bakery["store_id"] = bakery["store_id"].str.strip()
    return bakery

def item_clean(bakery):
    bakery["item"] = bakery["item"].replace({"케익" : "케이크"})
    return bakery

#중복 제거
def drop_duplicates_clean(bakery):
    bakery = bakery.copy()   # ← 이 순간, 함수 안의 'data'는 원본과 완전히 별개인 새 복사본이 됨
    bakery = bakery.drop_duplicates()   # 이후 모든 작업은 이 '복사본'에만 적용
    return bakery
```

```python
#pipeline
def preprocess(bakery, verbose=True):
    
    def log_step(data, step_name):
        if verbose:
            print(f"[{step_name}] 행 수: {len(data)}")
        return data
    
    bakery_clean = (
        bakery
            .pipe(bakery_copy).pipe(log_step, "bakery_copy") # orders_clean copy 생성 함수
            .pipe(revenue_clean).pipe(log_step, "revenue_clean") #revenue 음수 결측으로 처리
            .pipe(revenue_na_clean).pipe(log_step, "revenue_na_clean") #revenue 결측 계산값(unit_price * quantity)처리
            .pipe(date_str_parsing).pipe(log_step, "date_str_parsing") #날짜 파싱
            .pipe(store_id_clean).pipe(log_step, "store_id_clean") #store_id 문자열 처리
            .pipe(item_clean).pipe(log_step, "item_clean") #item 문자열 처리
    )
    return bakery_clean

bakery_clean = preprocess(bakery)
bakery_clean.head()
```

```python
duplicates = bakery_clean[bakery_clean.duplicated(keep=False)]   # 원본+복사본 모두 포함해서 보여줌
print(duplicates)
```

```python
def compare_summary(raw, clean, raw_name="bakery", clean_name="bakery_clean"):
    print("=" * 60)
    print("1. 행 수 비교")
    print("=" * 60)
    print(f"{raw_name}: {len(raw):,}행")
    print(f"{clean_name}: {len(clean):,}행")
    print(f"차이: {len(raw) - len(clean):,}행 감소")
    
    print()
    print("=" * 60)
    print("2. 컬럼별 결측 비율(%) 비교")
    print("=" * 60)
    missing_compare = pd.concat({
        raw_name: (raw.isna().sum() / len(raw) * 100).round(2),
        clean_name: (clean.isna().sum() / len(clean) * 100).round(2)
    }, axis=1)
    print(missing_compare)
    
    print()
    print("=" * 60)
    print("3. 중복 건수 비교")
    print("=" * 60)
    print(f"{raw_name}: {raw.duplicated().sum()}건")
    print(f"{clean_name}: {clean.duplicated().sum()}건 {(clean.duplicated().sum() / len(clean) *100).round(1)}%")

compare_summary(bakery, bakery_clean)
```

```python
#매장별 매출 순위(정제 후)

revenue_by_store = bakery_clean.groupby("store_id")["revenue"].sum().sort_values(ascending=False)
revenue_by_store_display = revenue_by_store.map(lambda x: f"{x:,}원")

print(f"bakery_clean 기준 \n {revenue_by_store_display}")

# 매장별 매출 순위(정제 전)
revenue_by_store_original = (bakery_clean[~bakery_clean["revenue_was_missing"]].groupby("store_id")["revenue"].sum().sort_values(ascending=False))
revenue_by_store_original_display = revenue_by_store_original.map(lambda x: f"{x:,}원")
print(f"\n bakery_clean 기준 \n {revenue_by_store_original_display}")
```

```python
compare = pd.concat({
    "원본 총 매출": revenue_by_store_original,
    "정제 후 총 매출": revenue_by_store
}, axis=1)

compare["총 매출 차이"] = compare["정제 후 총 매출"] - compare["원본 총 매출"]
compare["원본 순위"] = compare["원본 총 매출"].rank(ascending=False).astype("Int64")
compare["정제 후 순위"] = compare["정제 후 총 매출"].rank(ascending=False).astype("Int64")

# 표시용: 쉼표 포맷 적용
compare_display = compare.copy()
for col in ["원본 총 매출", "정제 후 총 매출", "총 매출 차이"]:
    compare_display[col] = compare_display[col].map(lambda x: f"{x:,.0f}" if pd.notna(x) else "-")

compare_display
```

```python
#월별 최고 매출 현황
pivot_original = (
    bakery_clean[~bakery_clean["revenue_was_missing"]]
    .pivot_table(index="store_id", columns="month", values="revenue", aggfunc="sum")
)

pivot = (bakery_clean.pivot_table(index="store_id", columns="month", values="revenue", aggfunc="sum"))

print("=== 원본(음수 및 NA값 제외) 기준 월별 최고 매출 매장 현황===")
for month in pivot_original.columns:
    top_store = pivot_original[month].idxmax()
    top_revenue = pivot_original[month].max()
    print(f"{month}월 최고 매출 매장: {top_store}, 금액: {top_revenue:,.0f}원")


print(f"\n=== revenue 정제 후 기준 월별 최고 매출 매장 현황 ===")
for month in pivot.columns:
    top_store = pivot[month].idxmax()
    top_revenue = pivot[month].max()
    print(f"{month}월 최고 매출 매장: {top_store}, 금액: {top_revenue:,.0f}원")
```
