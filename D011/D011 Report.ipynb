```python
import requests
import time
import random
import pandas as pd
from bs4 import BeautifulSoup
```

```python
tokyo_results = []

for page in range(1, 6):
    url = f"https://tabelog.com/kr/tokyo/rstLst/{page}"
    response = requests.get(url, timeout = 5)
    soup = BeautifulSoup(response.text, "html.parser")

    for name_tag in soup.select("div.list-rst__body"):
        rank = name_tag.select_one("i.list-rst__rank-badge-contents")
        store_name = name_tag.select_one("a.cpy-rst-name")
        stat = name_tag.select_one("div.cpy-area-genre")
        star = name_tag.select_one("span.list-rst__rating-val")
        info = name_tag.select_one("ul.list-rst__search-word")
        tokyo_results.append({
            "rank": rank.get_text(strip=True) if rank else None,
            "store_name": store_name.get_text(strip=True) if store_name else None,
            "stat": stat.get_text(strip=True) if stat else None,
            "star": star.get_text(strip=True) if star else None,
            "info": info.get_text(separator=" / ", strip=True) if info else None,
        })

        time.sleep(random.uniform(1, 2))

tokyo = pd.DataFrame(tokyo_results)
```

```python
# 원본 copy
tokyo_clean = tokyo.copy()
osaka_clean = osaka.copy()
kyoto_clean = kyoto.copy()
fukuoka_clean = fukuoka.copy()
hokkaido_clean = hokkaido.copy()
```

```python
# store_name 중복 검증
print(f"tokyo 업체명 중복: {tokyo_clean['store_name'].duplicated().sum()}건")
print(f"osaka 업체명 중복: {osaka_clean['store_name'].duplicated().sum()}건")
print(f"kyoto 업체명 중복: {kyoto_clean['store_name'].duplicated().sum()}건")
print(f"fukuoka 업체명 중복: {fukuoka_clean['store_name'].duplicated().sum()}건")
print(f"hokkaido 업체명 중복: {hokkaido_clean['store_name'].duplicated().sum()}건")
```

```python
# rank, star dtype 변경
tokyo_clean["rank"] = tokyo_clean["rank"].astype(int)
tokyo_clean["star"] = tokyo_clean["star"].astype(float)

osaka_clean["rank"] = osaka_clean["rank"].astype(int)
osaka_clean["star"] = osaka_clean["star"].astype(float)

kyoto_clean["rank"] = kyoto_clean["rank"].astype(int)
kyoto_clean["star"] = kyoto_clean["star"].astype(float)

fukuoka_clean["rank"] = fukuoka_clean["rank"].astype(int)
fukuoka_clean["star"] = fukuoka_clean["star"].astype(float)

hokkaido_clean["rank"] = hokkaido_clean["rank"].astype(int)
hokkaido_clean["star"] = hokkaido_clean["star"].astype(float)
```

```python
# stat split
tokyo_clean[["location", "menu"]] = tokyo_clean["stat"].str.split(" / ", n=1, expand=True)
tokyo_clean["location"] = tokyo_clean["location"].str.split(" ", n=1).str[0]
tokyo_clean["location"] = tokyo_clean["location"].str.strip()
tokyo_clean["menu"] = tokyo_clean["menu"].str.strip()

osaka_clean[["location", "menu"]] = osaka_clean["stat"].str.split(" / ", n=1, expand=True)
osaka_clean["location"] = osaka_clean["location"].str.split(" ", n=1).str[0]
osaka_clean["location"] = osaka_clean["location"].str.strip()
osaka_clean["menu"] = osaka_clean["menu"].str.strip()

kyoto_clean[["location", "menu"]] = kyoto_clean["stat"].str.split(" / ", n=1, expand=True)
kyoto_clean["location"] = kyoto_clean["location"].str.split(" ", n=1).str[0]
kyoto_clean["location"] = kyoto_clean["location"].str.strip()
kyoto_clean["menu"] = kyoto_clean["menu"].str.strip()

fukuoka_clean[["location", "menu"]] = fukuoka_clean["stat"].str.split(" / ", n=1, expand=True)
fukuoka_clean["location"] = fukuoka_clean["location"].str.split(" ", n=1).str[0]
fukuoka_clean["location"] = fukuoka_clean["location"].str.strip()
fukuoka_clean["menu"] = fukuoka_clean["menu"].str.strip()

hokkaido_clean[["location", "menu"]] = hokkaido_clean["stat"].str.split(" / ", n=1, expand=True)
hokkaido_clean["location"] = hokkaido_clean["location"].str.split(" ", n=1).str[0]
hokkaido_clean["location"] = hokkaido_clean["location"].str.strip()
hokkaido_clean["menu"] = hokkaido_clean["menu"].str.strip()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

font_path = "C:/Windows/Fonts/malgun.ttf"
fm.fontManager.addfont(font_path)
plt.rcParams["font.family"] = fm.FontProperties(fname=font_path).get_name()

cities = {
    "tokyo": tokyo_clean, "osaka": osaka_clean, "kyoto": kyoto_clean,
    "fukuoka": fukuoka_clean, "hokkaido": hokkaido_clean
}

# 도시별 menu 비율(%) 계산해서 하나의 표로 합치기
menu_by_city = {}
for name, df in cities.items():
    all_menus = df["menu"].str.split(",").explode().str.strip()
    menu_by_city[name] = all_menus.value_counts(normalize=True) * 100

menu_matrix = pd.DataFrame(menu_by_city).fillna(0)

# 전체 평균 기준 상위 10개 카테고리만 선택 (너무 많으면 복잡해지므로)
top_categories = menu_matrix.mean(axis=1).sort_values(ascending=False).head(10).index
menu_matrix_top = menu_matrix.loc[top_categories]

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(menu_matrix_top, annot=True, fmt=".1f", cmap="YlOrRd", ax=ax)
ax.set_title("도시별 Menu 비중(%) 비교")
plt.tight_layout()
plt.show()
```

```python
info_by_city = {}
for name, df in cities.items():
    all_info = df["info"].str.split(" / ").explode().str.strip()
    info_by_city[name] = all_info.value_counts(normalize=True) * 100

info_matrix = pd.DataFrame(info_by_city).fillna(0)

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(info_matrix, annot=True, fmt=".1f", cmap="YlGnBu", ax=ax)
ax.set_title("도시별 Info 비중(%) 비교")
plt.tight_layout()
plt.show()
```
